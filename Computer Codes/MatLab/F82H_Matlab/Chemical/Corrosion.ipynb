```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
msize = 10
fsize = 14

# Constants and calculations
a_1 = 7.3
b_1 = -5.8
a_2 = 7.85
b_2 = -5.0
del_val = 0.4
k = 8.617333e-5

A_1 = 10**(a_1 + del_val)
A_2 = 10**(a_2 + del_val / 2)
Q_1 = -1000 * k * b_1 / 0.434
Q_2 = -1000 * k * b_2 / 0.434
```

```python
def d_1(T):
    return A_1 * np.exp(-Q_1 / (k * T))

def d_2(T):
    return A_2 * np.exp(-Q_2 / (k * T))
```

```python
t = np.arange(0, 3.5, 0.5)

# Plot Dissolution distance for Li
plt.figure('Dissolution distance for Li')
plt.plot(t, d_1(673) * t, 'r', linewidth=2, label='400 C')
plt.plot(t, d_1(773) * t, 'b', linewidth=2, label='500 C')
plt.plot(t, d_1(873) * t, 'k', linewidth=2, label='600 C')
plt.legend(loc='best', fontsize=msize)
plt.xlabel('Time [years]', fontsize=fsize, fontname='Times New Roman')
plt.ylabel('Dissolution Distance [m]', fontsize=fsize, fontname='Times New Roman')
plt.xlim([0, 3])
plt.ylim([0, 50])
plt.title('Dissolution of F/M steels in Li')
plt.gca().tick_params(labelsize=fsize)
plt.grid(True)
plt.show()

# Plot Dissolution distance for LiPb
plt.figure('Dissolution distance for LiPb')
plt.plot(t, d_2(673) * t, 'r', linewidth=2, label='400 C')
plt.plot(t, d_2(773) * t, 'b', linewidth=2, label='500 C')
plt.plot(t, d_2(873) * t, 'k', linewidth=2, label='600 C')
plt.legend(loc='best', fontsize=msize)
plt.xlabel('Time [years]', fontsize=fsize, fontname='Times New Roman')
plt.ylabel('Dissolution Distance [m]', fontsize=fsize, fontname='Times New Roman')
plt.xlim([0, 3])
plt.ylim([0, 500])
plt.title('Dissolution of F/M steels in LiPb')
plt.gca().tick_params(labelsize=fsize)
plt.grid(True)
plt.show()
```
