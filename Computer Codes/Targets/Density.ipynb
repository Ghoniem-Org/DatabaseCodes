```python
def W_rho(T):
    if 0 <= T <= 3695:
        rho = 19.25 - 2.66207e-4 * (T - 298) - 3.0595e-9 * (T - 298)**2 - 9.5185e-12 * (T - 298)**3
    elif 3695 < T <= 6000:
        rho = 16.267 - 7.679e-4 * (T - 3695) - 8.091e-8 * (T - 3695)**2
    return rho * 1e3
```

```python
# Define the temperature range
T = np.arange(300, 6001, 20)
n = len(T)
```

```python
# Calculate density for each temperature value
rho = np.array([W_rho(temp) for temp in T])

# Plot the data
fig = plt.figure('W Thermal Conductivity')
plt.plot(T, rho, 'k', linewidth=2)
plt.xlabel('Temperature [K]', fontsize=14, fontname='Times New Roman')
plt.ylabel(r'$\rho$ [Kg/m$^3$]', fontsize=14, fontname='Times New Roman')
plt.xlim([0, 6000])
plt.ylim([0, 25000])
plt.title('Density of pure W', fontsize=14)
ax = plt.gca()
ax.set_xticks(np.arange(0, 7000, 1000))
ax.set_yticks(np.arange(0, 26000, 5000))
ax.tick_params(labelsize=14)

plt.show()
```
