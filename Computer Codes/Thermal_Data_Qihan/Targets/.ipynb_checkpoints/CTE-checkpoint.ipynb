### "A.-A. Tavassoli, J.-W. Rensman, M. Schirra, K. Shiba, Materials design data for reduced activation martensitic steel type f82h, Fusion Engineering and Design 61 (2002) 617–628."

```python
# Define the temperature range and function for CTE
T = np.arange(300, 6001, 20)
C_TE = lambda T: 9.0955 + 4.6477e-3 * T - 1.2141e-6 * T**2
```

```python
# Calculate CTE for each temperature value
CTE = C_TE(T)
```

```python
# Plot the data
fig = plt.figure('W Thermal Conductivity')
plt.plot(T, CTE, 'k', linewidth=2)
plt.xlabel('Temperature [K]', fontsize=14, fontname='Times New Roman')
plt.ylabel('CTE [ppm]', fontsize=14, fontname='Times New Roman')
plt.xlim([300, 1000])
plt.ylim([0, 15])
plt.title('Linear Thermal Expansion Coefficient of F82H', fontsize=14)
ax = plt.gca()
ax.set_xticks(np.arange(300, 1100, 100))
ax.set_yticks(np.arange(0, 16, 2))
ax.tick_params(labelsize=14)

plt.show()
```

```python
inital_row = 1
title = "Thermal Expansion Coefficient of F82H"
reference = ["A.-A. Tavassoli, J.-W. Rensman, M. Schirra, K. Shiba, Materials design data for reduced activation martensitic steel type f82h, Fusion Engineering and Design 61 (2002) 617–628."]

OutputExcel(sheetname, title, reference, inital_row, [], [fig])
```

### "Eurofer MPH 2016"

```python
temperature = [20, 50, 100, 200, 300, 400, 500]
mean_value = [10.3, 10.5, 10.7, 11.2, 11.6, 11.9, 12.2]
instantaneous_value = [10.3, 10.6, 11.1, 11.9, 12.6, 13.2, 13.5]

df_expansion = pd.DataFrame({
    "Temperature (°C)": temperature,
    "Mean Value (10^-6/K)": mean_value,
    "Instantaneous Value (10^-6/K)": instantaneous_value
})
```

```python
fig2 = plt.figure()
plt.plot(df_expansion["Temperature (°C)"], df_expansion["Mean Value (10^-6/K)"], marker='o', linestyle='-', color='b', label='Mean (α_m)')
plt.plot(df_expansion["Temperature (°C)"], df_expansion["Instantaneous Value (10^-6/K)"], marker='o', linestyle='-', color='r', label='Instantaneous (α_i)')
plt.title('Mean and Instantaneous Linear Thermal Expansion Coefficients')
plt.xlabel('Temperature (°C)')
plt.ylabel('Thermal Expansion (10^-6/K)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
inital_row = 30
title = "Thermal Expansion Coefficient of Eurofer"
reference = [
    "ASTM E-1142 Standard Terminology Relating to Thermophysical Properties",
    "ASTM E-228 Standard Test Method for Linear Thermal Expansion of Solid Materials With a Push-Rod Dilatometer",
    "J. Koslosky: “Using Coefficients of Thermal Expansion in a Finite Element Model” https://caeai.com/blog/coefficients-thermal-expansion-finite-element-model",
    "F. Tavassoli: “Comparison of 316L(N) -IG and Eurofer for Early DEMO”. EFDA-WP13-MAT-02-01"
]

OutputExcel(sheetname, title, reference, inital_row, [df_expansion], [fig2])
```
